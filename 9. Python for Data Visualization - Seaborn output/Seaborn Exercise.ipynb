```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.histplot(titanic['fare'],kde=False,color='red',bins=30)
```

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow',hue='class')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,hue='class', palette='Set2')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
corr_matrix = titanic.select_dtypes(include=['number']).corr()
plt.figure(figsize=(10, 6))

sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title('corr_matrix')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(sns.histplot,'age')
```
